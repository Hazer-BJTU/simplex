```python
import numpy as np

loaded = np.load("./datacache/final_points.npz")
final_points = {key: loaded[key] for key in loaded.files}
print(final_points)
```

```python
import matplotlib.pyplot as plt
from matplotlib.collections import PathCollection

#plt.style.use('seaborn-v0_8')
plt.rcParams.update({
    'font.family': 'Times New Roman',
    'font.size': 12,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'axes.grid': False,
    'grid.alpha': 1,
    'axes.prop_cycle': plt.cycler(color=plt.cm.tab10.colors)
})

num_tasks = 4
num_categories = 5

label_names = ['W', 'N1', 'N2', 'N3', 'REM']
dataset = ['ISRUC-1', 'SHHS', 'MASS', 'Sleep-EDF']

fig, axs = plt.subplots(3, 4, figsize=(20, 12))
def print_figure(feature_type: str, row: int):
    for t in range(num_tasks):
        centers = []
        subaxs = axs[row, t]
        if row == 2:
            subaxs.set_xlabel(f'Task{t + 1} {dataset[t]}', fontsize=26)
        if row == 0 and t == 0:
            subaxs.set_ylabel(f'Untrained generator', fontsize=26, va='bottom', ha='center', multialignment='center', labelpad=15)
        if row == 1 and t == 0:
            subaxs.set_ylabel(f'Well-trained generator', fontsize=26, va='bottom', ha='center', multialignment='center', labelpad=15)
        if row == 2 and t == 0:
            subaxs.set_ylabel(f'True feature distribution', fontsize=26, va='bottom', ha='center', multialignment='center', labelpad=15)
        for c in range(num_categories):
            points = final_points[f'task{t}_class{c}_' + feature_type]
            center = np.mean(points, axis=0)
            centers.append(center)
            subaxs.set_xticklabels([])
            subaxs.set_yticklabels([])
            subaxs.scatter(
                points[:, 0],
                points[:, 1],
                label=label_names[c],
                s=7,
                alpha=0.5,
                zorder=1
            )
        centers = np.array(centers)
        if t == 2 or t == 3:
            centers = centers[[0, 1, 4, 2, 3]]
        subaxs.scatter(
            centers[:, 0],
            centers[:, 1],
            marker='x',
            s=10,
            alpha=1,
            c='#333333',
            linewidth=5,
            zorder=5,
            label='cluster center'
        )
        subaxs.scatter(
            centers[:, 0],
            centers[:, 1],
            marker='x',
            s=100,
            alpha=1,
            c='#333333',
            linewidth=3,
            zorder=5,
        )
        subaxs.plot(
            centers[:, 0],
            centers[:, 1],
            linestyle='--', 
            color='#333333', 
            alpha=1, 
            linewidth=1.5,
            zorder=5
        )
print_figure('untrained_gen_feature', 0)
print_figure('gen_feature', 1)
print_figure('feature', 2)

handles, labels = axs[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, 
           loc='upper center',
           bbox_to_anchor=(0.5, 1.05),
           ncol=6,
           frameon=True,
           shadow=True,
           fontsize=18,
           markerscale=5.0)

plt.tight_layout()
plt.savefig('./figure/feature_vector.pdf', bbox_inches='tight')
plt.show()
```
